# Notebook 2: Exploratory Data Analysis
This notebook explores the College Scorecard dataset to identify institutional patterns associated with high student loan repayment risk. Five charts are included, each followed by a business insight written for a non-technical stakeholder. Key themes: institution type, income level, geography, and post-graduation earnings all influence repayment outcomes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/projects/student-loan-default-predictor/data/cleaned_scorecard.csv')

# Chart 1: Repayment risk by institution type
control_labels = {1: 'Public', 2: 'Private Nonprofit', 3: 'For-Profit'}
df['control_label'] = df['CONTROL'].map(control_labels)

risk_by_control = df.groupby('control_label')['repayment_risk'].mean().sort_values(ascending=False)
print(risk_by_control)

risk_by_control.plot(kind='bar', color=['#d73027','#fc8d59','#91bfdb'], figsize=(8,5))
plt.title('Repayment Risk Rate by Institution Type')
plt.ylabel('Proportion of High Risk Institutions')
plt.xlabel('Institution Type')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('C:/projects/student-loan-default-predictor/outputs/chart1_risk_by_control.png')
plt.show()
print("Chart saved.")
```

## Finding 1: Institution Type is a Strong Predictor of Repayment Risk

For-profit institutions show a repayment risk rate exceeding 50% â€” more than 4x higher than 
public institutions (approximately 12%). This suggests that loan servicers should apply 
heightened monitoring protocols to borrowers from for-profit schools at the point of 
enrollment, not after repayment struggles emerge.

```python
# Chart 2: Pell Grant percentage vs repayment risk
plt.figure(figsize=(8,5))
plt.scatter(
    df[df['repayment_risk']==0]['PCTPELL'],
    df[df['repayment_risk']==0]['COMPL_RPY_1YR_RT'],
    alpha=0.3, color='blue', label='Low Risk', s=10
)
plt.scatter(
    df[df['repayment_risk']==1]['PCTPELL'],
    df[df['repayment_risk']==1]['COMPL_RPY_1YR_RT'],
    alpha=0.3, color='red', label='High Risk', s=10
)
plt.xlabel('Pell Grant Recipients (%)')
plt.ylabel('1-Year Repayment Rate')
plt.title('Pell Grant Rate vs Repayment Performance')
plt.legend()
plt.tight_layout()
plt.savefig('C:/projects/student-loan-default-predictor/outputs/chart2_pell_vs_repayment.png')
plt.show()
print("Chart 2 saved.")
```

The majority of institutions fall below the 10% default rate threshold, but a meaningful tail extends beyond it. This right-skewed distribution suggests that default risk is concentrated in a specific subset of institutions making targeted early intervention far more cost-effective than broad policy changes.

```python
# Chart 3: Distribution of Repayment Risk
plt.figure(figsize=(9,6))
plt.hist(df['COMPL_RPY_3YR_RT'].dropna(), bins=30, color='#4575b4', edgecolor='white')
plt.axvline(x=df['COMPL_RPY_3YR_RT'].mean(), color='red', linestyle='--', linewidth=2, label='Average Repayment Rate')
plt.title('Distribution of 3-Year Repayment Rates Across Institutions', fontsize=14)
plt.xlabel('3-Year Repayment Rate')
plt.ylabel('Number of Institutions')
plt.legend()
plt.tight_layout()
plt.savefig('C:/projects/student-loan-default-predictor/outputs/chart3_repayment_distribution.png')
plt.show()
print("Chart 3 saved.")
```

Most institutions cluster around average repayment rates, but a significant number fall well below the mean indicating a concentrated group of high-risk institutions. Loan servicers should prioritize outreach to borrowers from institutions in the bottom quartile of repayment performance, where default risk is highest.

```python
# Chart 4: Top 15 States by Average Repayment Risk
plt.figure(figsize=(10,7))
state_risk = df.groupby('STABBR')['repayment_risk'].mean().sort_values(ascending=False).head(15)
state_risk.plot(kind='barh', color='#d73027', edgecolor='white')
plt.title('Top 15 States by Average Repayment Risk Rate', fontsize=14)
plt.xlabel('Proportion of High Risk Institutions')
plt.ylabel('State')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('C:/projects/student-loan-default-predictor/outputs/chart4_top_states.png')
plt.show()
print("Chart 4 saved.")
```

Repayment risk is not evenly distributed across the country certain states show a disproportionately high concentration of at-risk institutions. This geographic pattern suggests that state-level policy differences, local labor markets, and institution types all play a role. Federal loan servicers should consider state as a risk factor when building early intervention models.



```python
# Chart 5: Median Earnings vs Repayment Risk
plt.figure(figsize=(9,6))
sns.regplot(data=df, x='MD_EARN_WNE_P10', y='COMPL_RPY_3YR_RT',
            scatter_kws={'alpha':0.4, 'color':'#4575b4'},
            line_kws={'color':'red'})
plt.title('Median Earnings vs 3-Year Repayment Rate', fontsize=14)
plt.xlabel('Median Earnings 10 Years After Enrollment (USD)')
plt.ylabel('3-Year Repayment Rate')
plt.tight_layout()
plt.savefig('C:/projects/student-loan-default-predictor/outputs/chart5_earnings_vs_repayment.png')
plt.show()
print("Chart 5 saved.")
```
